```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

n = 600
x = np.random.rand(n)
m = 20
h = 30
M1 = np.zeros(m)
D1 = np.zeros(m)
for i in range(20):
  z = i*h
  M1[i] = np.mean(x[z:z+h])
  D1[i] = np.var(x[z:z+h], ddof=1)
print(M1)
print(D1)
```

```python
plt.plot(x)
plt.show()
```

```python
M1_med = np.median(M1)
D1_med = np.median(D1)
print('Мат.ожидание: ', M1_med)
print('Дисперсия: ', D1_med)
```

```python
plt.plot(M1)
plt.plot([M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med, M1_med])
plt.show()
```

```python
plt.plot(D1)
plt.plot([D1_med, D1_med,  D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med, D1_med,])
plt.show()
```

```python
M1_k = []
D1_k = []
for i in range(m):
  if M1[i]>M1_med:
    M1_k += [1]
  else: M1_k +=[0]
  if D1[i]>D1_med:
    D1_k += [1]
  else: D1_k +=[0]
```

```python
kD = 0
kM = 0
for i in range(m-1):
  if M1_k[i]!=M1_k[i+1]:
    kM+=1
  if D1_k[i]!=D1_k[i+1]:
    kD+=1
```

```python
print('Количество серий для мат.ожидания: ', kM)
print('Количество серий для дисперсии: ', kD)
if (kM>6 and kM<15):
  print('Опираясь на исследование мат.ожидания - Гипотеза о наличии тренда отвергается и процесс СТАЦИОНАРЕН при уровне значимости = 0.1 и 0.05')
else:
  print('Опираясь на исследование мат.ожидания - Гипотеза о наличии тренда принимается и процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.1 и 0.05')
if (kD>6 and kD<15):
  print('Опираясь на исследование дисперсии - Гипотеза о наличии тренда отвергается и процесс СТАЦИОНАРЕН при уровне значимости = 0.1 и 0.05')
else:
  print('Опираясь на исследование дисперсии - Гипотеза о наличии тренда принимается и процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.1 и 0.05')
```

```python
m_n = 0
d_n = 0
for i in range(m-1):
  for j in range(i+1, m):
    if M1[i]>M1[j]:
      m_n += 1
    if D1[i]>D1[j]:
      d_n += 1
```

```python
print('Количество инверсий для мат.ожидания: ', m_n)
if (m_n>69 and m_n<120):
  print('Опираясь на исследование мат.ожидания - процесс СТАЦИОНАРЕН при уровне значимости = 0.1')
else:
  print('Опираясь на исследование мат.ожидания - процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.1')
if (m_n>64 and m_n<125):
   print('Опираясь на исследование мат.ожидания - процесс СТАЦИОНАРЕН при уровне значимости = 0.05')
else:
  print('Опираясь на исследование мат.ожидания - процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.05')
```

```python
print('Количество инверсий для дисперсии: ', d_n)
if (d_n>69 and d_n<120):
  print('Опираясь на исследование дисперсии - процесс СТАЦИОНАРЕН при уровне значимости = 0.1')
else:
  print('Опираясь на исследование дисперсии - процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.1')
if (d_n>64 and d_n<125):
  print('Опираясь на исследование дисперсии - процесс СТАЦИОНАРЕН при уровне значимости = 0.05')
else:
  print('Опираясь на исследование дисперсии - процесс НЕСТАЦИОНАРЕН при уровне значимости = 0.05')
```
