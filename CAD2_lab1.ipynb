```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.fft
import matplotlib.ticker as ticker

A0 = 3
f0 = 60
beta = 90
t0 = 0.2
fi = m.pi/6
T = 0.5
N = int((0.5 - 0) / 0.002)
sigx = A0/2
epsi = np.random.normal(0, sigx, N)
ti = [0]
si = [0]
```

```python
for i in range(1, N):
  ti += [0 + i*0.002]
  si += [A0*np.exp(-(beta ** 2) * (ti[i] - t0) ** 2) * np.cos(2*m.pi*f0*(ti[i] - t0) + fi)]
```

```python
plt.figure(figsize=(10,5)) #шум
plt.plot(epsi + si)
plt.show()
```

```python
plt.plot(si) #полезный сигнал
```

```python
x = epsi + si
ACHx = scipy.fft.rfft(x)
ACHsi = scipy.fft.rfft(si) #описание в частотной области
plt.plot(abs(ACHsi))
```

```python
plt.plot(abs(ACHx))
```

```python
for i in range(len(ACHx)): #фильтрация
  if (i>15 and i<50):
    ACHx[i] = ACHx[i]
  else:
    ACHx[i] = 0
ACHx = np.array(ACHx)
plt.plot(ACHx)
```

```python
outx = scipy.fft.irfft(ACHx) #из частотной во временную(выход)
plt.plot(outx)
```
