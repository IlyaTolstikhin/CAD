```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
 
n = 450
z = 2*np.random.rand(n) - 1
mx = 0
dx = 1
mz = np.mean(z)
dz = np.std(z)
 
X = (dx/dz)*(z-mz)+mx
 
plt.plot(X)
plt.show()
```

```python
tau = 5
R = np.zeros(n//tau)
m = 0
for i in range(0, n-1, tau):
  for j in range(0, n-i, tau):
    R[m] += X[j+i]*X[j]
  R[m]=R[m]/(n-i) 
  m+=1
```

```python
plt.plot(R)
plt.show()
```

```python
tau = 10
R1 = np.zeros(n//tau)
m = 0
for i in range(0, n-1, tau):
  for j in range(0, n-i, tau):
    R1[m] += X[j+i]*X[j]
  R1[m]=R1[m]/(n-i) 
  m+=1
```

```python
plt.plot(R1)
plt.show()
```

АКФ - экспонента

```python
t0 = np.arange(0, 450)
dy = 3
Y = np.zeros(n)
R0 = dy * np.exp(-t0)
E0 = np.exp(-t0)
Q = (1 - E0**2) ** (0.5)
Y[1] = dy * X[1]
for i in range(1, n):
  Y[i] = E0[i]*Y[i-1] + dy*Q[i]*X[i] 
```

```python
plt.plot(Y)
plt.show()
```

```python
tau = 5
R = np.zeros(n//tau)
m = 0
for i in range(0, n-1, tau):
  for j in range(0, n-i, tau):
    R[m] += Y[j+i]*Y[j]
  R[m]=R[m]/(n-i) 
  m+=1
```

```python
plt.plot(R)
plt.show()
```

```python
tau = 10
R1 = np.zeros(n//tau)
m = 0
for i in range(0, n-1, tau):
  for j in range(0, n-i, tau):
    R1[m] += Y[j+i]*Y[j]
  R1[m]=R1[m]/(n-i) 
  m+=1
```

```python
plt.plot(R1)
plt.show()
```

```python
import scipy.stats as sts
r = R/R[0]
 
a = 0.18
Zr = 0.5*np.log((1+r[1:])/(1-r[1:]))
delta = sts.norm.ppf(1-a/2)/((n-3)**0.5)
print(r[1:])
print('Нижние границы доверительного интервала')
print(Zr-delta)
print('Верхние границы доверительного интервала')
print(Zr+delta)
```

```python
plt.figure()
plt.plot(r[1:])
plt.scatter(range(len(Zr-delta)), Zr-delta)
plt.scatter(range(len(Zr-delta)), Zr+delta)
plt.show()
```
