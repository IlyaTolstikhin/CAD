```python
import numpy as np
import math
import scipy.stats as sts
import matplotlib.pyplot as plt
n = 1000
R = np.zeros(n)
x = np.random.rand(n)
m = np.mean(x)
m2 = np.mean(x**2)
s = np.var(x, ddof=1)
X = x - m # центрируем
k = math.floor(1.72 * (n ** (1/3)))
for i in range(n-1):
  for j in range(n-i):
    R[i] += X[j+i]*X[j]
  R[i]=R[i]/(n-i) 
plt.hist(x, bins = k, range=(0, 1))
plt.show()
 
print('Оценка математического ожидания: ', m)
print('Оценка среднего квадрата:', m2)
print('Оценка дисперсии: ', s)
```

```python
plt.plot(R[::4])
plt.show()
```

```python
Ro = R/np.max(R)
plt.plot(Ro[:3])
plt.show()
```

```python
for i in range(1, n):
  if R[i]<=Ro[0]:
    t0 = i
    break
print('Рассчетное время корреляции =', t0)
```

```python
errm = np.zeros(n)
errm2 = np.zeros(n)
errs = np.zeros(n)
 
for i in range(n):
  x = np.random.rand(n)
  m1 = np.mean(x)
  m21 = np.mean(x**2)
  s1 = np.var(x, ddof=1)
  errm[i] = (m1-m)**2
  errm2[i] = (m21-m2)**2
  errs[i] = (s1-s)**2
 
epsm = (np.mean(errm))**0.5
epsm2 = (np.mean(errm2))**0.5
epss = (np.mean(errs))**0.5
```

```python
print('Среднеквадратическая ошибка оценки мат. ожидания: ', epsm)
print('Среднеквадратическая ошибка оценки квадрата среднего: ', epsm2)
print('Среднеквадратическая ошибка оценки дисперсии: ', epss)
print('Доверительный интервал оценки мат. ожидания: (', m/(1+2*epsm),';',m/(1-2*epsm),')')
print('Доверительный интервал оценки квадрата среднего: (', m2/(1+2*epsm2),';',m2/(1-2*epsm2),')')
print('Доверительный интервал оценки дисперсии: (', s/(1+2*epss),';',s/(1-2*epss),')')
```
