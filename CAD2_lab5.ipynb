```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.fft
```

```python
b = 60
w0 = np.pi * 40
v0 = 0
t0 = -0.05
tn = 0.05
N = int((tn - t0) / 0.002)
epsi = [] # шум
ti = []
si = []
```

```python
epsi = np.random.normal(1, 2, N)
```

```python
for i in range(N):
  ti += [-0.05 + i*0.002]
  si += [np.exp(-(b * ti[i])** 2) * np.cos(w0 * ti[i] + v0)]
```

```python
plt.figure(figsize=(10,5)) #чистый сигнал
plt.plot(si)
plt.show()
```

```python
y = np.zeros(N)
for i in range(N):
  y[i] = np.array([epsi[i] + si[i]])
plt.figure(figsize=(10,5)) # входной сигнал
plt.plot(y)
plt.show()
```

```python
ACHy = scipy.fft.rfft(y)
plt.plot(abs(ACHy))
```

```python
ACHsi = scipy.fft.rfft(si) # АЧХ чистого
plt.plot(abs(ACHsi))
```

```python
ACHepsi = scipy.fft.rfft(epsi) # АЧХ шума
plt.plot(abs(ACHepsi))
```

```python
ACHsi = np.array(ACHsi)
ACHepsi = np.array(ACHepsi)
```

```python
H = ACHsi/(ACHsi+ACHepsi)
plt.plot(abs(H))
```

```python
ACHout = ACHy*H
plt.plot(abs(ACHout))
```

```python
out = scipy.fft.irfft(ACHout)
plt.figure(figsize=(10,5)) #чистый сигнал
plt.plot(out)
plt.show()
```
