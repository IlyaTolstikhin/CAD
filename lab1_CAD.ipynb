```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
n = 10000
l = 5
s = [0 for i in range(n)]
for i in range(n):
  rd = random.random()
  r = 1
  k = 0
  while (r>math.exp(-l)):
    r = r*rd
    k+=1
  s[i] = int(k)
print(s)
m = math.floor(1.72 * (n ** (1/3)))
plt.hist(s, bins = m, range=(0, 50))
plt.show()
```
