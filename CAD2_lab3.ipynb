```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.fft
```

```python
A0 = 1
w = 2*m.pi*40
beta = 60
t0 = - 0.03
tn = 0.03
fi = 0
N = int((tn - t0) / 0.0002)
sigx = A0/50
ti = [t0]
y = [A0*m.cos(w*ti[0] + fi)]
```

```python
for i in range(1, N):
  ti += [t0 + i*0.002]
  y += [A0*m.cos(w*ti[i] + fi)]
```

```python
plt.plot(y)
```

```python
ACHy = scipy.fft.rfft(y)
plt.plot(abs(ACHy))
```

```python
# амплитудно-модулированный сигнал
y1 = [m.exp(-(beta ** 2) * (ti[0] ** 2))*m.cos(w*ti[0] + fi)]
for i in range(1, N):
  y1 += [m.exp(-(beta ** 2) * (ti[i] ** 2))*m.cos(w*ti[i] + fi)]
```

```python
plt.plot(y1)
```

```python
ACHy1 = scipy.fft.rfft(y1)
plt.plot(abs(ACHy1))
```

```python
# частотно-модулированный сигнал
y2 = [A0*m.cos(ti[0]*ti[0] + fi)]
for i in range(1, N):
  y2 += [A0*m.cos(ti[i]*ti[i] + fi)]
```

```python
plt.plot(y2)
```

```python
ACHy2 = scipy.fft.rfft(y2)
plt.plot(abs(ACHy2))
```

```python
# фазо-модулированный сигнал
y3 = [A0*m.cos(w*ti[0] + m.sin(ti[0]))]
for i in range(1, N):
  y3 += [A0*m.cos(w*ti[i] + m.sin(ti[i]))]
```

```python
plt.plot(y3)
```

```python
ACHy3 = scipy.fft.rfft(y3)
plt.plot(abs(ACHy3))
```

```python
fig, axs = plt.subplots(nrows= 2 , ncols= 2 )
fig. suptitle('АЧХ сигналов')
axs[0, 0].plot(abs(ACHy), label = 'гармоника')
axs[0, 1].plot(abs(ACHy1), label = 'АМ')
axs[1, 0].plot(abs(ACHy2), label = 'ЧМ')
axs[1, 1].plot(abs(ACHy3), label = 'ФМ')
fig.set_figheight(10)
fig.set_figwidth(15)
plt.show()
```
