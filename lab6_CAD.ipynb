```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
 
n = int(7.5/(0.002))
x = np.random.rand(n)
m = np.mean(x)
X = x - m
R = np.zeros(n-1)

for i in range(n-1):
  for j in range(n-i):
    R[i] += X[j+i]*X[j]
  R[i]=R[i]/(n-i)
```

```python
plt.plot(R)
plt.show()
```

```python
S = np.real(fft(R))
```

```python
plt.plot(S)
plt.show()
```
